```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stockdata.csv')

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10, 6))

plt.plot(df['Date'], df['MSFT'], label='MSFT', color='b', linestyle='-', linewidth=2)
plt.plot(df['Date'], df['IBM'], label='IBM', color='c', linestyle='-', linewidth=2)
plt.plot(df['Date'], df['SBUX'], label='SBUX', color='r', linestyle='-', linewidth=2)
plt.plot(df['Date'], df['AAPL'], label='AAPL', color='g', linestyle='-', linewidth=2)
plt.plot(df['Date'], df['GSPC'], label='GSPC', color='m', linestyle='-', linewidth=2)

plt.xlabel('Data')
plt.ylabel('Valore')
plt.title('Andamento delle Azioni')

plt.legend(loc='lower right')

plt.show()
```
