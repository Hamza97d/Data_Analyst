```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stockdata.csv')

aapl_data_prime_20 = df['AAPL'].head(20)
date_data_prime_20 = pd.to_datetime(df['Date']).head(20)

plt.figure(figsize=(10, 6))

plt.plot(date_data_prime_20, aapl_data_prime_20, marker='o', linestyle='--', color='red', markersize=8, markerfacecolor='black', linewidth=2)

plt.xlabel('Data')
plt.ylabel('Valore')
plt.title('Azioni Apple')
plt.xticks(rotation='vertical')

plt.show()
```
