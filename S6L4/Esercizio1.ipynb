```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stockdata.csv')  

msft_data = df['MSFT']
date_data = pd.to_datetime(df['Date'])  

plt.plot(date_data,msft_data)
plt.xlabel('Data')
plt.ylabel('Valore Azione MSFT')
plt.title('Valore Azioni MSFT')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stockdata.csv')

msft_data_prime_5_righe = df['MSFT'].head(5)
date_data_prime_5_righe = pd.to_datetime(df['Date']).head(5)

plt.figure(figsize=(10, 6)) 
plt.plot(date_data_prime_5_righe, msft_data_prime_5_righe, linestyle='-', color='c', label='MSFT')
plt.xlabel('Data')
plt.ylabel('Valore Azione MSFT')
plt.title('Valore Azioni MSFT - Prime 5 righe')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stockdata.csv')

msft_data_ultime_5_righe = df['MSFT'].tail(5)
date_data_ultime_5_righe = pd.to_datetime(df['Date']).tail(5)

plt.figure(figsize=(10, 6)) 
plt.plot(date_data_ultime_5_righe, msft_data_ultime_5_righe, linestyle='-', color='m', label='MSFT')
plt.xlabel('Data')
plt.ylabel('Valore Azione MSFT')
plt.title('Valore Azioni MSFT - Ultime 5 righe')
plt.show()
```
