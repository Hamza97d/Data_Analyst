```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
#passeggeri divisi per classe
conteggio_passeggeri = titanic['class'].value_counts()

sns.barplot(x=conteggio_passeggeri.index, y=conteggio_passeggeri.values)

plt.title('Numero di Passeggeri per Classe di Imbarco')
plt.xlabel('Classe di Imbarco')
plt.ylabel('Numero di Passeggeri')
plt.show()
#morti/vivi divisi per classe
conteggio_sopravvissuti = titanic.groupby(['class', 'alive']).size().unstack()

conteggio_sopravvissuti.plot(kind='bar', stacked=True)

plt.title('Numero di Passeggeri Sopravvissuti/Morti per Classe')
plt.xlabel('Classe')
plt.ylabel('Numero di Passeggeri')
plt.legend(title='Vivo/Morto')
plt.show()

#distribuzione tariffe
sns.histplot(titanic['fare'])

plt.title('Distribuzione delle Tariffe')
plt.xlabel('Tariffa')
plt.ylabel('Frequenza')

plt.show()
```
