```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')



conta_ponti = titanic['deck'].nunique()
print('I ponti sono:',conta_ponti)


colonne_con_valori_nulli = titanic.isnull().sum()[titanic.isnull().sum() > 0]
print('valori nulli:\n',colonne_con_valori_nulli)


sns.lmplot(x='fare', y='age', data=titanic)
plt.show()
```
