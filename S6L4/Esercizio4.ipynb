```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('election.csv')

totale_voti = df[['Coderre', 'Bergeron', 'Joly']].sum()

# Grafico Voti totali per candidato
totale_voti.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Voti Totali per ogni candidato')
plt.xlabel('Candidato')
plt.ylabel('Voti Totali')
plt.xticks(rotation=0)

plt.show()

# Grafico Votanti per distretto
df.plot(x='district_id', y='total', kind='bar', color='green')
plt.title('Numero di Votanti per Distretto')
plt.xlabel('Distretto')
plt.ylabel('Numero di Votanti')
plt.xticks(rotation=90)

plt.show()



primi_4_distretti = df.head(4)

voti_df = primi_4_distretti.set_index('district')[['Coderre', 'Bergeron', 'Joly']]

# Grafico a barre appaiato 
voti_df.plot(kind='bar', figsize=(12, 6))
plt.title('Voti per i Candidati nei Primi 4 Distretti (Appaiato)', fontsize=16)
plt.xlabel('Distretto', fontsize=14)
plt.ylabel('Numero di Voti', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Candidato')
plt.savefig('Voti_per_candidato_nei_primi_4_distretti_appaiato.png', dpi=300)
plt.show()

# Grafico a barre impilato 
voti_df.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Voti per i Candidati nei Primi 4 Distretti (Impilato)', fontsize=16)
plt.xlabel('Distretto', fontsize=14)
plt.ylabel('Numero di Voti', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Candidato')
plt.savefig('Voti_per_candidato_nei_primi_4_distretti_impilato.png', dpi=300)
plt.show()


```
