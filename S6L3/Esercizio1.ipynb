```python
import pandas as pd
import numpy as np

dataset = 'wine.csv'
df = pd.read_csv(dataset)

colonne = df.select_dtypes(include="number").columns


statistiche = df.describe()

mediana = df[colonne].median()
moda = df[colonne].mode().iloc[0]  
media = df[colonne].mean()
quartili = df[colonne].quantile([0.25, 0.5, 0.75])

print("Statistiche:")
print(statistiche)

print("\nMediana:")
print(mediana)

print("\nModa:")
print(moda)

print("\nMedia:")
print(media)

print("\nQuartili:")
print(quartili)
```
